```python
# Create two tensors containing random values between 0 and 1 with shape [5, 300].
tensor_1 = tf.random.uniform([5, 300], minval=0, maxval=1)
tensor_2 = tf.random.uniform([5, 300], minval=0, maxval=1)

tensor_1, tensor_2
```

```python
# Multiply the two tensors you created in 3 using matrix multiplication.
result1 = tf.matmul(tensor_1, tensor_2, transpose_b=True) # (5x5)
result2 = tf.matmul(tensor_1, tensor_2, transpose_a=True) # (300x300)

result3 = tf.matmul(tensor_2, tensor_1, transpose_b=True) # (5x5)
result4 = tf.matmul(tensor_2, tensor_1, transpose_a=True) # (300x300)

result1, result2, result3, result4
```

```python
# Multiply the two tensors you created in 3 using dot product.
result5 = tf.tensordot(tf.transpose(tensor_1), tensor_2, axes=1)  # 300x5 . 5x300 -> 300x300
result6 = tf.tensordot(tensor_1, tf.transpose(tensor_2), axes=1)  # 5x300 . 300x5 -> 5x5

result7 = tf.tensordot(tf.transpose(tensor_2), tensor_1, axes=1)  # 300x5 . 5x300 -> 300x300
result8 = tf.tensordot(tensor_2, tf.transpose(tensor_1), axes=1)  # 5x300 . 300x5 -> 5x5

result5, result6, result7, result8
```

# NN Regression - Exercises

```python
def plot_predictions(
    train_in, train_out, 
    test_in, test_out,
    preds,
    title='Plot-Predictions',
    figsize=(10, 7),
):
    plt.figure(figsize=figsize)
    # Plot training data in blue
    plt.scatter(train_in, train_out, c="b", label="Training")
    # Plot test data in green
    plt.scatter(test_in, test_out, c="g", label="Testing")

    assert isinstance(preds, dict)

    # Plotting predictions
    for label, y_preds in preds.items(): plt.scatter(test_in, y_preds, label=label)

    plt.title(title)
    plt.legend();
```

```python
# Create your own regression dataset and build, fit a model to it.
import matplotlib.pyplot as plt

# Create features
X = np.arange(-100, 100, 4)

# Create labels
y = np.arange(-90, 110, 4)

# Split data into train and test sets
X_train = X[:40] # first 40 examples (80% of data)
y_train = y[:40]

X_test = X[40:] # last 10 examples (20% of data)
y_test = y[40:]

# Visualize it
plt.figure(figsize=(10, 7))
# Plot training data in blue
plt.scatter(X_train, y_train, c='b', label='Training data')
# Plot test data in green
plt.scatter(X_test, y_test, c='g', label='Testing data')
# Show the legend
plt.legend();
```

```python
# Try building a neural network with 4 Dense layers and fitting it to your own regression dataset, how does it perform?

tf.random.set_seed(42)

# Create a model using the Sequential API
model1 = tf.keras.Sequential([
  tf.keras.layers.Dense(1),
  tf.keras.layers.Dense(1),
  tf.keras.layers.Dense(1),
  tf.keras.layers.Dense(1)
])

# Compile the model
model1.compile(
    loss=tf.keras.losses.mae, # mae is short for mean absolute error
    optimizer=tf.keras.optimizers.SGD(), # SGD is short for stochastic gradient descent
    metrics=["mae"]
  )

# Fit the model
# model.fit(X, y, epochs=5) # this will break with TensorFlow 2.7.0+
history_1 = model1.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=5)
```

```python
import pandas as pd

def plot_losses(histories):
    assert isinstance(histories, dict)

    plt.figure(figsize=(20, 14))

    for label, history in histories.items():    
        losses = history.history['loss']
        plt.plot(losses, label=label)

    plt.title('Losses')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend();




```
